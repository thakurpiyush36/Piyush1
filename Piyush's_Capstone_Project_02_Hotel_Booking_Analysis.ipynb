- `arrival_date_year`: Year of arrival
- `arrival_date_month`: Month of arrival
- `arrival_date_week_number`: Week number of the year for arrival
- `arrival_date_day_of_month`: Day of arrival
- `stays_in_weekend_nights`: Number of weekend nights stayed or booked
- `stays_in_week_nights`: Number of week nights stayed or booked
- `adults`: Number of adults
- `children`: Number of children
- `babies`: Number of babies
- `country`: Country of origin in ISO 3166–1 alpha-3 format

##We'll check if there is any null value or missing values in our dataset


```python
df_Hotels.isnull()
```

```python
# We'll sort the null value as per the columns
df_Hotels.isnull().sum().sort_values(ascending=False)
```

### Replacing the null values

```python
#In the given dataset we have 112593 null values in 'company' column, so we will replcae all the null values with 0
df_Hotels['company'] = df_Hotels['company'].fillna(0)
df_Hotels['company'].head(20)
```

```python
# In the "agent" column we have got 16340 null values and we'll replace them with 0
df_Hotels['agent'] = df_Hotels['agent'].fillna(0)
df_Hotels['agent'].head(20)
```

```python
# In the "country" column we have got 488 null values and we'll replace them with "OTH" as the code for others
df_Hotels['country'] = df_Hotels['country'].fillna('OTH')
df_Hotels['country'].head(50)
```

```python
#In the "children" column we have got 4 null values and we'll replace them with 0
df_Hotels['children'] = df_Hotels['children'].fillna(0)
```

## Deleting the rows with incorrect or false data

```python
# There are several rows where the overall count of guests including adult, babies and children are 0
df_Hotels[(df_Hotels.adults+df_Hotels.babies+df_Hotels.children)==0].shape
```

```python
# In total there are 180 such rows where the overall count of the guests are 0, we'll drop all such rows
df_Hotels = df_Hotels.drop(df_Hotels[(df_Hotels.adults+df_Hotels.babies+df_Hotels.children)==0].index)
```

#Exploratory Data Analysis and Visualization


##1. What is the ratio of different type of hotels?
In the given dataset there are two types of hotels "City Hotel" and "Resort Hotel". The following pie chart visually represents the relative share in percentages for each type of hotel

```python
# Count the occurrences of each room type
room_type_counts = df_Hotels['hotel'].value_counts()

# Plot a pie chart
plt.figure(figsize=(4, 4))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.0f%%', startangle=180)
plt.title('Division of Hotel Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

The pie chart reveals that 66% of the hotels in the dataset are 'City Hotels' and the other 34% are 'Resort Hotels'.



##2. In which month we have the arrival of maximum numbers of guest?
The plot below shows us the number of arrivals for the bookings as per months of the year.


```python
monthly_arrival = df_Hotels['arrival_date_month'].value_counts()

# Reorder the months to plot them in chronological order
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_arrival = monthly_arrival.reindex(months_order)

plt.figure(figsize=(10, 6))
plt.plot(monthly_arrival.index, monthly_arrival.values, marker='o', linestyle='-', color='b')
plt.xlabel('Month')
plt.ylabel('Number of Arrivals')
plt.title('Monthly Arrivals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 From the above graph it can be observed that generally the months of November to March are off-peak periods for hoteliers. Between June and August, attendance is at its peak, it is the high season.

## 3. What is average daily room rate for all the months?
From the below plot we can see the average daily rate of rooms for all the months across year and this will help us to find when the rooms will be cheaper.

```python
import matplotlib.pyplot as plt

# Group the data by 'arrival_date_month' and calculate average daily rate (ADR)
monthly_adr = df_Hotels.groupby('arrival_date_month')['adr'].mean()

# Reorder the months to plot them in chronological order
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_adr = monthly_adr.reindex(months_order)

# Plot the average daily rate (ADR) over time (by month)
plt.figure(figsize=(10, 6))
plt.plot(monthly_adr.index, monthly_adr.values, marker='o', linestyle='-', color='b')
plt.xlabel('Month')
plt.ylabel('Average Daily Rate (in $)')
plt.title('Average Daily Rate of Rooms Over Different Months')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

From the graph above we can see that the prices of the rooms are pn the higher side in the peak season which is June to August, and prices are relatively lower in off-peak season that is November to March.

##4. Countries where we had the most bookings
The below chart represent the top 10 countries in terms of bookings counts

```python
top10_countries = df_Hotels.groupby('country')['country'].count().sort_values(ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(10, 6))
colors = sns.color_palette('bright', n_colors=len(top10_countries))
sns.barplot(x=top10_countries.index, y=top10_countries.values,hue = top10_countries.index,palette=colors)
plt.xlabel('Country')
plt.ylabel('Number of Guests')
plt.title('Top 10 Countries in Terms of Booking Counts')

# Show the plot
plt.show()
```

 From the above graph it can be observed that the most guests came from PRT which the code of Portugal, followed by GBR which is the Great Britain, followed bt France and so on.


##5. How many bookings got cancelled ?

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
ax = sns.countplot(x='hotel', hue='is_canceled', data=df_Hotels, palette='bright')

plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.title('Booking Cancellations by Hotel Type')
plt.legend(title='Booking Status', loc='upper right', labels=['Not Canceled', 'Canceled'])

plt.show()
```

In this plot Blue shaded area represents "Booking Not Cancelled" and Orange shaded area represents "Booking Cancelled" and it is observed that more than 30000 booking were cancelled in City hotels followed by more than 10000 in Resort hotels.

##6. What are the top 5 market segment in terms of number of bookings?

```python
top_segments = df_Hotels['market_segment'].value_counts().head(5)
plt.figure(figsize=(10,6))
plt.title("Segments wise booking")
sns.barplot(x = top_segments.index,y = top_segments.values,hue = top_segments ,palette = "bright")
plt.xticks(rotation = 45)
plt.xlabel("Market Segment")
plt.ylabel("No. of Bookings")
plt.show()
```

 Most of the market_segment used  Online TA and Offline TA/TO channels which is travel agency and tour operator.Followed by Groups and direct channel.

##7. What is the percentage of the repeated guests?


```python
repeated_guests = df_Hotels['is_repeated_guest'].value_counts()
plt.figure(figsize=(10,6))
plt.title("Percentage of repeated guests")
custom_labels = ['Not Repeated Guest', 'Repeated Guest']
plt.pie(repeated_guests,labels=custom_labels, autopct='%1.0f%%')
plt.title(" Percentgae (%) of repeated guests")
plt.show()
```

The percentage of repeat guests are relatively less which is only 3% ,
In order to retain more guests, hotels will have to work towards their services and have to take guest's feedback seriously.
